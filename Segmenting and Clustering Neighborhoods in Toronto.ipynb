The data is scraped from Wikipeda using beautiful soup

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

And read into a data frame

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

Unfortunately the data frame is quite messy and the entries are badly formatted. So each entry of the data frame needs to be extracted, cleaned, and put into a new data frame. The cleaning is mainly extracting the first 3 letters as the postcode, splitting the rest of the name on the opening parenthesis to separate borough from neighbourhood, and then getting rid of the 'Not Assigned' entries in the appropriate manner (deleting or replacing).

```python
pstcd = list()
brgh = list()
nbhd = list()
count = 0

for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        count += 1
        pstcd.append(df[j][i][:3])
        brgh.append(df[j][i][3:].split('(')[0])
        if brgh[len(brgh)-1] != 'Not assigned':
            nbhd.append(df[j][i][3:].split('(')[1][:-1].replace(" /", ","))
        else:
            brgh.pop()
            pstcd.pop()

nbhd[:5]
```

```python
dic = {'Postcode': pstcd, 'Borough': brgh, 'Neighbourhood': nbhd}

dft = pd.DataFrame(data = dic)

dft['Borough'] = dft['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})

dft.head(10)
```

Now we being working with the geospatial data, which is then merged with the above data frame

```python
lat_lon = pd.read_csv('Geospatial_Coordinates.csv')
lat_lon.head()
```

```python
lat_lon.rename(columns={'Postal Code':'Postcode'},inplace=True)
df3 = pd.merge(dft,lat_lon,on='Postcode')
df3.head(10)
```

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4.head(10)
```

A map can now be created by layering the data frame co-ordinates onto a folium map

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

We can begin the data clustering

```python
k=5
toronto_clustering = df4.drop(['Postcode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)

df4.head()
```

And finally display the clustered data points on a new map

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'],
                                            df4['Longitude'],
                                            df4['Neighbourhood'],
                                            df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
